```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt 
import math
from math import *
```

```python
npoints = 1000
t = np.linspace(-20.0,20.0,npoints)
n = np.array([1,2,1])
r0 = np.array([-1,2,-3])

#  Given line 
xline = -1 + 1*t
yline =  2 + 2*t
zline = -3 + 1*t

# Data for a three-dimensional line
xlines = np.zeros((6,npoints))
ylines = np.zeros((6,npoints))
zlines = np.zeros((6,npoints))

# get the angle between two vectors a and b
def angle(a,b):
    return acos(   np.dot(a,b)/(   sqrt(np.dot(a,a))*sqrt(np.dot(b,b))    )      )

# rotate vectors v on the xy-plane to the plane with normal vector n about the origin
def rotateToPlane(v,n):
    #z-axis, the normal vector for the xy plane
    z =  np.array([0,0,1])
    #obtain angles for rotation
    x =  np.array([1,0,0])
    N = np.cross(z,n)
    a = angle(x,N)
    b = angle(z,n)
    c = pi/6
    
    #rotate about x-axis
    Rza = np.array([[cos(a), -1*sin(a),0],[sin(a), cos(a),0],[0,0,1]])
    Rxb = np.array([[1,0,0],[0, cos(b), -1*sin(b)],[0, sin(b), cos(b)]])
    R = np.matmul(Rza, Rxb)
    
    # uncomment to check if the z-zxis is parallel to n after transformatoin
    # print(np.matmul(R, z))
    return np.matmul(R, v)
    
# generate the unit vectors of the six lines 30 degrees apart in the x-y plane
# then rotate them into the plane that is perpendiclular to n
for i in range(6):
    theta = 2*pi*i/12.0
    v = np.array([cos(theta), sin(theta),0])
    v = rotateToPlane(v,n)
    # uncomment to test if all the generated lines are perpedicular to the given line
    # print(np.dot(v,n) <= 0.000001)
    xlines[i] = -1 + t*v[0]
    ylines[i] =  2 + t*v[1]
    zlines[i] = -3 + t*v[2]
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
w=2
ax.set_xlim(r0[0]-w, r0[0]+w)
ax.set_ylim(r0[1]-w, r0[1]+w)
ax.set_zlim(r0[2]-w, r0[2]+w)
ax.plot3D(xline, yline, zline, 'black')
for i in range(6):
    ax.plot3D(xlines[i], ylines[i], zlines[i], 'blue')

xa = degrees(angle([1,0,0],[1,2,0]))
ya = degrees(angle([1,2,0],[1,2,1]))
ax.view_init(30,70) #choose the angle that you want to see the 3D graph.
```
