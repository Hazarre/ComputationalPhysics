```python
import numpy as np
from math import *
import matplotlib.pyplot as plt 
```

```python
#boundary condition 
y0 = 5-5

def diffy(x,y):
    return (2*x-6*x*y)/(1+x**2) 


def ay(x):
    return (1-1/(1+x**2)**3)/3.0
```

```python
#define variables
h =0.1
X = np.arange(0.0,100, h)
npoints = 1000

Y = np.zeros(npoints)
Y_analytical = np.zeros(npoints)

Y[0] = 0
Y_analytical[0] = 0

# numerically approximate the function that satisfies our differential equation


for i in range(1,npoints):
    
    Y[i] = (Y[i-1]+h*diffy(X[i-1],Y[i-1]))
    Y_analytical[i] = ay(X[i]) 
    
```

```python
# plot all the series in the range of 0-0.5
plt.plot(X,Y)
plt.plot(X,Y_analytical,c='orange')
plt.show()
```

The Euler forward method is a good estimation when the step size h is small. See below for the significant discrepancies when h becomes larger. 

```python
#define variables
h = 0.5
npoints = 100
up = npoints*h
X = np.arange(0.0,up,h)


Y = np.zeros(npoints)
Y_analytical = np.zeros(npoints)

Y[0] = 0
Y_analytical[0] = 0

# numerically approximate the function that satisfies our differential equation


for i in range(1,npoints):
    
    Y[i] = (Y[i-1]+h*diffy(X[i-1],Y[i-1]))
    Y_analytical[i] = ay(X[i]) 
    
```

```python
# plot all the series in the range of 0-0.5
plt.plot(X,Y)
plt.plot(X,Y_analytical,c='orange')
plt.show()
```
