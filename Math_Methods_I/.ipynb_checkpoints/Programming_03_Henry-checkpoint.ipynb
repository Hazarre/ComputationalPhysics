Question 1 (20 points). Plot the result of problem 2.17.26 as a function of theta (0,2π),

using the numpy functions “abs()” and/or “conj()” and confirm that it matches your ana-
lytical solution.



The -i numerator is modified to +i

```python
import numpy as np
import matplotlib.pyplot as plt 
import math
```

```python
#To save you some time, in python an i is denoted as 1j
npoints = 1000
x = np.linspace(0,2*np.pi,npoints)

def compute(theta):
    num = 2*np.exp(theta*1j) + 1j
    den = 1j*np.exp(theta*1j) + 2
    return np.abs(num/den) # np.sqrt(num*np.conj(num)/(den*np.conj(den)))


def analyze(theta):
    y = np.sin(theta)
    num = 5 + 4*y
    den = 5 - 4*y
    return np.sqrt(num/den)

ypython =  np.full(npoints, (0+0j) )
yanalytical = np.full(npoints, (0+0j) )

for index in range(npoints):
    ypython[index] = compute(x[index]) 
    yanalytical[index] = analyze(x[index])
    
    
```

```python
plt.plot(x,ypython,x,yanalytical)
plt.show()
```

Since the analytical and the computed graph overlap with each other, yes the computer caculated result matches my analytical answer.
