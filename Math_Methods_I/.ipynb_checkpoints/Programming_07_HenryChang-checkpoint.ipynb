```python
import numpy as np
from math import *
import matplotlib.pyplot as plt 
```

```python
#define variables
k = 8.99*10**9
e = 1.6*10**(-19)
hb = 1.054*10**(-34)
me = 9.11*10**(-31)

h = 10**(-14)
E = -2.1725527*10**(-18)
npoints = 100000

r = np.arange(0.0, npoints*h, h)
u = np.zeros(npoints)
du = np.zeros(npoints)
duu = np.zeros(npoints)

# initial conditions
u[0] = 0
du[0] = 1
duu[0] = 0 

for i in range(1,npoints):
    du[i] = du[i-1]+ h*duu[i-1]
    u[i]  = u[i-1] + h*du[i-1]
    duu[i] = -2*me/hb**2*(E+k*e**2/r[i])*u[i]
```

```python
#You will need to plot u versus r in order to see how the solution looks. Choose the
#following axes limits: Horizontal axis: 0 − 10−9 m, Vertical axis: 0 − 10−10
plt.ylim(top=10**(-10))
plt.xlim(right=10**(-9))
plt.plot(r,u)
plt.show()
```

u blows up to infinity when E <= -2.1725527*10**(-18)
