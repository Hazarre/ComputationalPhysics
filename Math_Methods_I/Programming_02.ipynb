# Power series 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#define variables
npoints = 1000
nterms = 4

x = np.linspace(0,0.5,npoints) #general npoints numbers between 0 and 0.5 with equal spacing

# Define the power series S(n+1) = S(n)+variable^n, calculate and return an array of series [0, S1 ... SN]  
def series(variable,N):
    sum = np.zeros(N+1)
    for i in range(N):
        sum[i+1]=sum[i]+variable**i
    return sum[N]
```

```python
#generate npoints numbers between 0 and 0.5
result1 = np.zeros(npoints)
result2 = np.zeros(npoints)
result3 = np.zeros(npoints)
actual = np.zeros(npoints)

for index in range(npoints):
     # power series S(n+1) = S(n) + (0.5/999)^n, n begins from zero
    result1[index] = series(x[index],2) # approximate with the series above until n = 2
    result2[index] = series(x[index],4) # approximate with the series above until n = 4
    result3[index] = series(x[index],6) # approximate with the series above until n = 4
    actual[index] = 1/(1-x[index]) # the actual function y = 1/(1-x)
```

```python
# plot all the series in the range of 0-0.5
plt.plot(x,result1,x,result2,x,result3,x,actual)
plt.show()
```

The graph showed that, in the range of [0,0.5]:<br>
The power series with degree 2 approximates the function poorly.
The power series with degree 4 approximates the function well.
The power series with degree 6 approximates the function very well.

So the power series converges to the function very quickly as it approximates the function from poorly to very well by raising a couple degrees. 

```python
from math import *
x = np.linspace(0,2*pi,npoints) #general npoints numbers between 0 and 2pi with equal spacing

# Define the power series for sin(x)
def myseries(variable,N):
    sum = np.zeros(N+1)
    for i in range(N):
        #display(i)
        sum[i+1]=sum[i]+variable**(2*i)*(-1)**i/factorial(2*i)
        #display(sum[i+1])
    return sum[N]

result1 = np.zeros(npoints)
result2 = np.zeros(npoints)
result3 = np.zeros(npoints)
actual = np.zeros(npoints)

for index in range(npoints):
    result1[index] = myseries(x[index],5) # approximate with the series above until n = 5
    result2[index] = myseries(x[index],7) # approximate with the series above until n = 7
    result3[index] = myseries(x[index],8) # approximate with the series above until n = 8
    actual[index] = cos((x[index]))

plt.plot(x,result1,x,result2,x,result3,x,actual)
plt.show()
```


The graph showed that, in the range of [0,2pi]:<br>
The power series with degree 5 approximates the function poorly.
The power series with degree 7 approximates the function well.
The power series with degree 8 approximates the function very well.
The power series with degree 9 will approximate the function accurately.

We need at least 8 terms (degree 7) of the power series to approx. the function reasonably well.
