```python
import numpy as np
import matplotlib.pyplot as plt 
#plank constant
hbar =(6.62607004*(10**(-34)))/(2*np.pi)
```

```python
def series1(N):
    sum=np.zeros(N+1)
    for i in range(N):
        sum[i+1]=sum[i]+1
    return sum

def series2(N):
    sum=np.zeros(N+1)
    for i in range(N):
        sum[i+1]= sum[i]+0
    return sum
```

```python
maxN = 10000
sum1 = series1(maxN)
sum2 = series2(maxN)
ns   = np.zeros(maxN+1)
for i in range(maxN+1):
    ns[i] = i
```

```python
plt.plot(ns,sum1,ns,sum2)
plt.show()
```
