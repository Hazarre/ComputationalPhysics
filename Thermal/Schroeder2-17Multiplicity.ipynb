```python
import numpy as np
import matplotlib.pyplot as plt

def omega(z, N):
    # z = qa/q 
    # (1-z) = qb/q
    return (4*z*(1-z))**N

Ns = [1, 10 , 100, 1000, 10000]
```

```python
for i in range(len(Ns)):
    N = Ns[i]
    plt.plot(z, omega(z, N), label= ('n=%s' % Ns[i]))
    plt.ylabel('multiplicity')
    plt.xlabel('z')
plt.legend()
plt.show()
```
