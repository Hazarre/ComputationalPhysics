```python
import numpy as np
import matplotlib.pyplot as plt 
from math import * 
```

```python
def piOverX(X):
    return X/pi

def fSeries(N,X):
    temp = np.zeros(X.shape) 
    for n in range(1,N+1):
        temp += -2*cos(n*pi)*np.sin(n*X)/pi/n
    return temp    
```

```python
lb = -pi # lowerbound
ub = pi #upperbound
npoints = 1000 
spacing = (ub-lb)/npoints
X = np.arange(lb, ub + spacing, spacing)


# initialize functions to be plot
f1 = np.copy(X)
f2 = np.copy(X)
f3 = np.copy(X)
```

```python
f1 = piOverX(X)
f2 = fSeries(1,X)
f3 = fSeries(2,X)
f4 = fSeries(3,X)
f5 = fSeries(4,X)

```

```python
plt.plot(X,f1, X, f2, X, f3, X, f4, X, f5)
plt.figure(dpi=1000)
plt.legend()
plt.show()
```
