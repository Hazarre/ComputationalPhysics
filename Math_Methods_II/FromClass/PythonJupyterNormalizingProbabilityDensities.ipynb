# Normalizing and Plotting Probability Densities

As usual we begin by loading the libraries that we will need. In this case they are scipy, a scientific programming library, numpy, for access to numerical routines, and matplotlib, which gives us access to various plotting techniques.

```python
import scipy.integrate as integrate
import numpy as np
import matplotlib.pyplot as plt
```

As you can see, this loads the integrate function from scipy. It will allow us to do definite integrals numerically. 

For example, consider a straight line of slope 1 that goes from $x=0$ to $x=1$. This is just a triangle of base one and height one and so its area should be $1/2$. Let's check this numerically:

```python
integrate.quad(lambda x: x,0,1)
```

The .quad tells integrate that I want to use the method of "quadratures", which is a particular way of doing numerical integrals. The first argument "lambda x:" tells python that it should treat $x$ like a variable in a function; there is a wealth of computer science behind programming functions and you can look up "lambda calculus" if you'd like to learn more about it. Here you can just think of this as a simple way to define a function. 

After "lambda x:" appears the specific function of this example, $x$ gives the straight line of slope 1. The next argument is where the integral should begin, here 0, and after that where it should end, which is at 1 in this example. 

The integrate function returns two numbers. The first is the numerical result of the integration. The second is the error in its value. Because this is a numerical routine, you do have to be a bit careful about errors. If we integrate $x^{20}(1-x)^{24}$ over this same interval, we get an error that's as big as the value of the integral:

```python
integrate.quad(lambda x: x**20*(1-x)**24,0,1)
```

It will often be convenient to name the outputs of integrate.quad and access the result of the integral as the 0th element of the outputs:

```python
outs=integrate.quad(lambda x: x,0,1)
outs[0]
```
